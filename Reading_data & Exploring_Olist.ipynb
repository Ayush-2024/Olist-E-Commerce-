```python
# Average Delivery Time Analysis
delivery_df = df_order.select('order_id','order_purchase_timestamp','order_delivered_customer_date')
```

```python
delivery_df.show()
```

```python
delivery_df = delivery_df.withColumn(
    "delivery_time",
    datediff(col("order_delivered_customer_date"), col("order_purchase_timestamp"))
).orderBy(col("delivery_time").desc())
delivery_df.display()
```
